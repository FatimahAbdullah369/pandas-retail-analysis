```python
import pandas as pd

# Create a mock dictionary of retail data with all numbers included
data = {
    'Transaction_ID': [101, 102, 103, 104, 105, 106],
    'Product': ['Laptop', 'Headphones', 'Coffee Maker', 'Running Shoes', 'Desk Lamp', 'Headphones'],
    'Category': ['Electronics', 'Electronics', 'Home', 'Apparel', 'Home', 'Electronics'],
    'Price': [1200, 150, 80, 120, 30, 150],
    'Quantity': [1, 2, 1, 2, 3, 1]
}

# Turn the dictionary into a Pandas DataFrame (table)
df = pd.DataFrame(data)

# View the first few rows of our table
df.head()
```

```python
# Check how many rows and columns we have (Rows, Columns)
print("Data Shape:", df.shape)

# Check data types and see if there are missing values (like SQL's schema info)
df.info()
```

```python
# Multiply Price column by Quantity column
df['Total_Revenue'] = df['Price'] * df['Quantity']

# View the updated table
df
```

```python
# Filter rows where Category is 'Electronics'
electronics_df = df[df['Category'] == 'Electronics']

# View the filtered table
electronics_df
```

```python
# Filter rows where Total_Revenue is greater than 100
expensive_sales = df[df['Total_Revenue'] > 100]

# View the filtered table
expensive_sales
```

```python
# Filter for Home category AND quantity > 1
home_bulk_sales = df[(df['Category'] == 'Home') & (df['Quantity'] > 1)]

# View the filtered table
home_bulk_sales
```

```python
# Group by Category and sum up the Total_Revenue column
category_revenue = df.groupby('Category')['Total_Revenue'].sum()

# View the result
print("Total Revenue by Category:")
print(category_revenue)
```

```python
# Group by Category and find the mean (average) of the Price column
category_avg_price = df.groupby('Category')['Price'].mean()

# View the result
print("\nAverage Price by Category:")
print(category_avg_price)
```

```python
# Group by Category and count the Transaction_IDs
category_counts = df.groupby('Category')['Transaction_ID'].count()

# View the result
print("\nTransaction Count by Category:")
print(category_counts)
```
